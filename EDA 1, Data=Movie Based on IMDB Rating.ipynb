```python
fig= plt.figure(figsize=(15,10))
plt.bar(data.head(10)['title'],data.head(10)['runtime'])
plt.xlabel("runtime",fontsize=30)
plt.ylabel("title",fontsize=28)
plt.show()
```

```python
fig= plt.figure(figsize=(15,10))
plt.bar(data.tail(10)['title'],data.tail(10)['runtime'])
plt.xlabel("runtime",fontsize=30)
plt.ylabel("title",fontsize=28)
plt.show()
```

```python
#RATING ON YEAR BASIS 
sns.stripplot(x="votes", y="rating", data=data.head(10), jitter=True);
plt.title("RATING BASED ON VOTES")
plt.show()
```

```python
#METASCORE BASED ON YEAR BASIS 
sns.stripplot(x="rating", y="metascore", data=data, jitter=True, color="green");
plt.title("METASCORE BASED ON RATING")
plt.show()
```

```python

sns.countplot(data=data.head(10),x="metascore",hue='genre')
plt.xticks(rotation=90)
plt.show()
```
